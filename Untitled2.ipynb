```python
import pandas_datareader.data as pdr
from datetime import datetime, timedelta

end_date = datetime.now()
start_date = end_date - timedelta(days=1825)

df_goog = pdr.DataReader('JBSS3.SA', 'yahoo', start_date, end_date)
df_goog.tail()
```

```python
df_goog[['High', 'Low']].plot(figsize=(15, 4), title='JBS ', grid=True)
```

```python
df_goog['simple_return'] = (df_goog['Adj Close']) / (df_goog['Adj Close'].shift(1)) - 1
print (df_goog['simple_return'])
```

```python
avg_returns_d = df_goog['simple_return'].mean()
avg_returns_d
#media diaria de retorno
```

```python
avg_returns_a = df_goog['simple_return'].mean() *1000
avg_returns_a
```

```python
print (str(round(avg_returns_a, 5)* 100) + '%')
```
