```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
```

```python
tickers = ['VVAR3.SA', 'JBSS3.SA', 'LCAM3.SA', 'OIBR3.SA']
myData = pd.DataFrame()
for t in tickers:
    myData[t] = wb.DataReader (t,data_source='yahoo',start ='2019-01-01' )['Adj Close']
```

```python
myData.info()
```

```python
myData.head()
```

```python
myData.tail()
```

```python
myData.iloc[0]
```

```python
(myData / myData.iloc[0] * 100). plot(figsize=(15,6));
plt.show()
```

```python
returns = (myData / myData.shift(1)) - 1
returns
```

```python
weights = np.array([0.25,0.25,0.25,0.25])
#definindo o peso de cada ativo na sua carteira
```

```python
np.dot(returns,weights)
#calcula o produto de um vetor ou matriz, multiplicando o peso de cada ao ao seu respectivo retorno
```

```python
annual_returns = returns.mean() * 250 
annual_returns
```

```python
np.dot(annual_returns, weights)
```

```python
pfolio_1 = str(round(np.dot(annual_returns, weights), 5)*100) +'%'
#CALCULANDO PORCENTAGEM RETORNO
```

```python
pfolio_1
```

```python
weights_2 = np.array ([0.4,0.4,0.1,0.1])
```

```python
pfolio_2 = str(round(np.dot(annual_returns, weights_2), 5)*100) +'%'
print (pfolio_1)
```
